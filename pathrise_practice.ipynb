```python
# length_of_job_search

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('length_of_job_search').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# employment_status  vs highest_level_of_education

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['highest_level_of_education'].value_counts()
    for x_label, grp in df.groupby('employment_status ')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('employment_status ')
_ = plt.ylabel('highest_level_of_education')
```

```python
# @title Number of Interviews vs Number of Applications

import matplotlib.pyplot as plt
plt.scatter(df['number_of_interviews'], df['number_of_applications'])
plt.xlabel('Number of Interviews')
plt.ylabel('Number of Applications')
_ = plt.title('Number of Interviews vs Number of Applications')
```

```python
# duration_days

from matplotlib import pyplot as plt
df['duration_days'].plot(kind='hist', bins=20, title='duration_days')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
